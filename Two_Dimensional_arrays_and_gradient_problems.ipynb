## Problem 1: Linear Function

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-50, 50.1, 0.1) # 50.1 to include 50

# Calculate y = 1/2x + 1
y = 1/2 * x + 1

print("Elements of x:", x)
print("Elements of y:", y)
```

## Problem 2: Array Combination

```python
xy = np.column_stack((x, y))

print("Shape of combined array:", xy.shape)
print("Combined array elements:\n", xy)
```

## Problem 3: Find the gradient

```python
dy = np.diff(y) # Differences beetween adjacent y values
dx = np.diff(x) # Differences beetween adjacent x values
gradient = dy / dx

print("Gradient values:\n", gradient)
print("Length of gradient array:", len(gradient))
```

## Problem 4: Draw a graph

```python
import numpy as np
import matplotlib.pyplot as plt

# x for the gradient has one less value, so we trim x
x_gradient = x[:-1]

# Set up the plots
plt.figure(figsize=(12, 5))

# Plot for the original function(linear function)
plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st subplot
plt.plot(x, y, label='y = 0.5x + 1', color='green')
plt.title("Linear Function")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()

# Plot the Gradient
plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2st subplot
plt.plot(x_gradient, gradient, label='Gradient', color='blue')
plt.title("Gradient at Each x")
plt.xlabel("x")
plt.ylabel("Gradient")
plt.grid(True)
plt.legend()

# Show both plots
plt.tight_layout()
plt.show()
```

## Problem 5: Python functionalization

```python
import numpy as np
import matplotlib.pyplot as plt

def compute_gradient(function, x_range=(-50, 50.1, 0.1)):
  """ calculate the gradient using the amount change.
  Parameters:
  function: function
      A functions that returns ndarray 'y' when given ndarray 'x'
  x_range: tuple
      range like (start, stop, step) for np.arange()

  Returns:
  array_xy: ndarray, shape(n, 2)
      combined 'x' and 'y' values
  gradient: ndarray, shape(n - 1.)
      gradient (dy/dx) between adjacent points
  """
  x = np.arange(*x_range) # create x values
  y = function(x) # calculate 'y' values using the provided function
  dy = np.diff(y)
  dx = np.diff(x)
  gradient = dy / dx
  array_xy = np.column_stack((x, y))
  return array_xy, gradient

```

```python
# compute required functions
def function1(x):
  return x ** 2

def function2(x):
  return 2 * x ** 2 + 2 * x

def function3(x):
  return np.sin(x /12)

# compute gradients for all 3 functions
# for y = x^2
xy1, grad1 = compute_gradient(function1)

# for y = 2x^2 + 2x
xy2, grad2 = compute_gradient(function2)

# for y = sin(x/12) x from 0 to 50
xy3, grad3 = compute_gradient(function3, x_range=(0, 50.1, 0.1)) #as per given range
```

```python
# Draw graphs
def function_gradient_plot(xy, gradient, title_prefix):
  x = xy[:, 0]
  y = xy[:, 1]
  x_grad = x[:-1]

  plt.figure(figsize=(12, 5))

  #plot 1: Funtion
  plt.subplot(1, 2, 1)
  plt.plot(x, y, label=f'{title_prefix} Function', color='black')
  plt.xlabel('x')
  plt.ylabel('y')
  plt.title(f'{title_prefix} Function')
  plt.grid(True)
  plt.legend()

  #plot 2: Gradient
  plt.subplot(1, 2, 2)
  plt.plot(x_grad, gradient, label=f'{title_prefix} Gradient', color='purple')
  plt.xlabel('x')
  plt.ylabel('Gradient')
  plt.title(f'{title_prefix} Gradient')
  plt.grid(True)
  plt.legend()

  plt.tight_layout()
  plt.show()

#plot all three
function_gradient_plot(xy1, grad1, "y = x")
function_gradient_plot(xy2, grad2, "y = 2x + 2x")
function_gradient_plot(xy3, grad3, "y = sin(x/12)")
```

## Problem 6: Find the minimum value

```python
def analyze_minimum(xy, gradient, func_label):
  y = xy[:, 1]
  x = xy[:, 0]

  min_index = np.argmin(y)
  min_y = y[min_index]
  min_x = x[min_index]

  print(f"\n--- {func_label} ---")
  print(f"Minimum y = {min_y:.4f} at x = {min_x:.4f}")

  #Show slope before and after, if they exist
  if 1 <= min_index < len(gradient):
    print(f"Slope before minimum: {gradient[min_index - 1]:.4f}")
    print(f"Slope after minimum: {gradient[min_index]:.4f}")
  else:
    print("Minimum is at the edge of the domain, cannot compute both slopes.")

# function 1: y = x^2
xy1, grad1 = compute_gradient(function1)
analyze_minimum(xy1, grad1, "y = x^2")

# function 2: y = 2x^2 + 2x
xy2, grad2 = compute_gradient(function2)
analyze_minimum(xy2, grad2, "y = 2x^2 + 2x")

# function 3: y = sin(x/12) x from 0 to 50
xy3, grad3 = compute_gradient(function3, x_range=(0, 50.1, 0.1))
analyze_minimum(xy3, grad3, "y = sin(x/12)")
```
